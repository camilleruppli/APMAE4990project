# Calculate distance in km using haversine function

```python
def radians(x):
    radian = np.arccos(-1)*x/180
    return radian
def radians_to_km(x):
    km = 111.045*180*x/np.arccos(-1)
    return km  
def haversine(lat1,long1,lat2,long2):
    haversine_dist = radians_to_km(np.arccos(np.cos(radians(lat1))*np.cos(radians(lat2))*np.cos(radians(long1)-radians(long2))+np.sin(radians(lat1))*np.sin(radians(lat2))))
    return haversine_dist
```

```python
testing_data['distance_in_km'] = np.zeros(testing_data.shape[0])
p_lat = testing_data['pickup_latitude'].values
p_long = testing_data['pickup_longitude'].values
d_lat = testing_data['dropoff_latitude'].values
d_long = testing_data['dropoff_longitude'].values
distance_in_km = testing_data['distance_in_km'].values
for i in range(testing_data['distance_in_km'].shape[0]):
    distance_in_km[i] = haversine(d_lat[i],d_long[i], d_lat[i], p_long[i])+haversine(d_lat[i],d_long[i], p_lat[i], d_long[i])
```

# Fill in all time features to match those in the training set

```python
import datetime
def weekdays(df,new_col_name,col_name):
    df[new_col_name]=np.zeros(df.shape[0])
    new_col=df[new_col_name].values
    info_col=df[col_name].values
    for i in range(df.shape[0]):
        new_col[i]=datetime.datetime.strptime(str(info_col[i]),'%Y-%m-%d  %H:%M:%S+00:00').strftime('%w')
    return df
```

```python
testing_data=weekdays(testing_data,'day_of_week','pickup_datetime')
```

```python
def weekend(df,new_col_name,col_name):
    df[new_col_name]=np.zeros(df.shape[0])
    new_col=df[new_col_name].values
    info_col=df[col_name].values
    for i in range(df.shape[0]):
        if info_col[i]!=0.0 and info_col[i]!=6.0:
            new_col[i]=1
    return df
```

```python
testing_data=weekend(testing_data,'weekday','day_of_week')
```

```python
def day_time_bin(df,new_col_name,col_name,bin_num):
    df[new_col_name]=np.zeros(df.shape[0])
    new_col=df[new_col_name].values
    info_col=df[col_name].values
    for i in range(df.shape[0]):
        new_col[i]=info_col[i]/bin_num
    return df
```

```python
testing_data=day_time_bin(testing_data,'day_binned','day_of_week',7)
```

```python
testing_data=day_time_bin(testing_data,'time_binned','day_hour',24)
testing_data.head()
```

```python
def day_time_info(df,new_col_name,day_col_name,time_col_name,bin_num_day,bin_num_hour):
    df[new_col_name]=np.zeros(df.shape[0])
    new_col=df[new_col_name].values
    time_col=df[time_col_name].values
    day_col=df[day_col_name].values
    for i in range(df.shape[0]):
        new_col[i]=(day_col[i]+time_col[i]/bin_num_hour)/bin_num_day
    return df
```

```python
testing_data=day_time_info(testing_data,'day_number','day_of_week','day_hour',7,24)
```

```python
def sine_cosine(df,cosine_col_name,sine_col_name,info_col_name):
    df[cosine_col_name]=np.zeros(df.shape[0])
    df[sine_col_name]=np.zeros(df.shape[0])
    info_col=df[info_col_name].values
    cosine_col=df[cosine_col_name].values
    sine_col=df[sine_col_name].values
    for i in range(df.shape[0]):
        cosine_col[i]=np.cos(2*np.pi*info_col[i])
        sine_col[i]=np.sin(2*np.pi*info_col[i])
    return df
```

```python
testing_data=sine_cosine(testing_data,'day_cosine','day_sine','day_number')
testing_data=sine_cosine(testing_data,'time_cosine','time_sine','time_binned')
```
